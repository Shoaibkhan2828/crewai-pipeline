```python
import pandas as pd

def clean_dataset(data):
    """
    A streamlined data-cleaning utility that transforms raw CSV datasets into structured, analysis-ready formats.

    What this function does:
    - Removes duplicate rows to ensure uniqueness.
    - Replaces all missing values with the placeholder string "Missing".
    - Normalizes column names by converting to lowercase and replacing spaces with underscores.
    - Detects columns containing the word 'date' and converts them to datetime format (when possible).

    Parameters:
        data (pd.Dataframe): The dataset to be cleaned.

    Returns:
        pd.DataFrame: Cleaned and formatted dataset ready for analysis or export.
    """
    # Remove duplicate rows
    data.drop_duplicates(inplace=True)

    # Fill missing values with placeholder
    data.fillna("Missing", inplace=True)

    # Normalize column names to lowercase with underscores
    data.columns = [col.strip().lower().replace(' ', '_') for col in data.columns]

    # Convert any column containing 'date' to datetime format
    for col in data.columns:
        if 'date' in col:
            try:
                data[col] = pd.to_datetime(data[col])
            except Exception:
                pass

    return data,

clean_dataset(data)
```

```python
from crewai.tools import tool
import pandas as pd

@tool("Data_Collection_Tool")
def get_data_insights():
    """
    This tool is used to present insights of a dataset.
    It loads a CSV file and returns a summary including:
    - Dataset shape (rows, columns)
    - Count of missing values per column
    - Data types of each column
    - First 5 sample records for quick inspection
    """
    global data
    file_path = "/content/sample_data/california_housing_test.csv"
    data = pd.read_csv(file_path)
    insights = {
        "Shape": data.shape,
        "Missing Values": data.isnull().sum().to_dict(),
        "Data Types": data.dtypes.astype(str).to_dict(),
        "Sample Records": data.head(5).to_dict(orient="records")
    }
    return f"The insights of dataframe are {insights}"
```

```python
from crewai.tools import tool
import pandas as pd

@tool("Data_Cleaning_Tool")
def clean_dataset() -> str:
    """
    A streamlined data-cleaning utility that transforms raw CSV datasets into structured, analysis-ready formats.

    What this function does:
    - Removes duplicate rows to ensure uniqueness.
    - Replaces all missing values with the placeholder string "Missing".
    - Normalizes column names by converting to lowercase and replacing spaces with underscores.
    - Detects columns containing the word 'date' and converts them to datetime format (when possible).

    Returns:
        str: A message indicating the completion of the task.
    """
    data.drop_duplicates(inplace=True)
    data.fillna("Missing", inplace=True)
    data.columns = [col.strip().lower().replace(' ', '_') for col in data.columns]

    for col in data.columns:
        if 'date' in col:
            try:
                data[col] = pd.to_datetime(data[col])
            except Exception:
                pass

    return "The dataset is cleaned"
```

```python
from dataclasses import KW_ONLY
from crewai import Agent, Task, Crew, LLM
from crewai_tools import CSVSearchTool
from crewai.tools import tool
from crewai_tools import CodeInterpreterTool
# Write content to a file in a specified directory
# llm = LLM(model="groq/deepseek-r1-distill-llama-70b",api_key=key)
from crewai import LLM

llm = LLM(
    model="gemini/gemini-2.5-pro",
    temperature=0.7,
    api_key = key
)
run_codes = CodeInterpreterTool(unsafe_mode=True)


# Create your CrewAI agents with role, main goal/objective, and backstory/personality
Data_Collection_Agent = Agent(
    role='Responsible for presenting the insights of a dataset.', # Agent's job title/function
    goal='Your main objective is that you will be given with a dataset you have to return the data insights like summary,etc from the dataset', # Agent's main objective
    backstory='You are an expert in understanding datasets ', # Agent's background/expertise
    llm=llm, # LLM that powers your agent
    verbose=True,
    tools = [get_data_insights] # Show agent's thought process as it completes its task

)

Data_Cleaning_Agent = Agent(
    role='Specialist in organizing, cleaning, and transforming raw datasets into structured, high-quality data.',
    goal='From the insights given ensure if all collected data is free from inconsistencies, duplicates, and inaccuracies to support reliable analysis and decision-making.',
    backstory='A detail-oriented data wrangler with deep expertise in preprocessing, normalization, and error detection across diverse datasets. Highly adept at using automation and logic to turn messy data into meaningful insights.',
    llm=llm,
    verbose=True,
    tools = [clean_dataset]
)
Data_Analysing_Agent = Agent(
    role= 'Data Analyst',
    goal='Analyze the cleaned dataset to identify key trends, patterns, and insights.',
    backstory='You are an experienced data analyst skilled in exploring datasets, identifying correlations, and present findings.',
    llm=llm,
    verbose=True,
)
Data_Plotting_Agent= Agent(
    role = 'Data Vizualizer',
    goal = 'Read the data given provided by Data Analyst and create a matplotlib chart from that data If you are given specific instructions on how to draw the chart then follow them, if not then write the code and the execute the code and save the image file',
    backstory= '"You an expert at analyzing and creating charts',
    llm=llm,
    verbose=True,
    tools = [run_codes]
    #allow_code_execution=True
)
```

```python

# Define your agents' tasks
Data_Collection_Task = Task(
    description='You will be given a CSV path. You have to invoke the tool given to you and provide insights',
    expected_output='A point wise desciption of the insights given by the tool output',
    agent=Data_Collection_Agent  # Assigned agent
)
Data_Cleaning_Task = Task(
    description='Process and structure collected data into a clean, tabular format. Save the results in a standardized CSV file for easy access and further analysis.',
    expected_output='It should be structured format.',
    agent=Data_Cleaning_Agent,
    dependencies=[Data_Collection_Agent])
Data_Analysis_Task = Task(
    description='Analyze the cleaned dataset (available in the `data` variable) to identify key trends, patterns, and insights. Focus on understanding the relationships between variables, and present your findings.',
    expected_output='A specific pointwize of the key trends, patterns, and insights found in the dataset.',
    agent=Data_Analysing_Agent,
    dependencies=[Data_Collection_Agent],
    max_rpm = 9000
)
Data_Plotting_Task= Task(
    description= 'Analyze the cleaned dataset (available in the `data` variable), Create a Chart and save at {/content/alifornia_housing_train.png}',
    expected_output='A image file',
    agent=Data_Plotting_Agent,
    dependencies=[Data_Analysing_Agent]
)
```

```python
# Create crew to manage agents and task workflow
crew = Crew(
    agents=[Data_Collection_Agent, Data_Cleaning_Agent, Data_Analysing_Agent,Data_Plotting_Agent], # Agents to include in your crew
    tasks=[Data_Collection_Task, Data_Cleaning_Task, Data_Analysis_Task,Data_Plotting_Task], # Tasks in execution order
    verbose=True
)

result = crew.kickoff(inputs={"any": "Get the data insights of dataset in csv_path:/content/sample_data/california_housing_test.csv "})
print(result)
```
